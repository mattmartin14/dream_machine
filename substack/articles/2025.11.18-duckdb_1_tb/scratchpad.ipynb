```python
import duckdb
cn = duckdb.connect()

f_path = f'/Volumes/xd1/data/*.parquet'
```

```python
cn.sql(f"""
    select sum(size)/1024.00/1024.00/1024.00/1024.00 as total_size_tb
    from read_blob('{f_path}')
""").show()
```

```python
cn.sql(f"select count(*) as row_cnt, sum(rand_val) as total_rand_val from read_parquet('{f_path}')").show()
```

```python
cn.sql(f"""
    select rand_dt, count(*) as row_cnt, sum(rand_val) as total_rand_val 
    from read_parquet('{f_path}')
    group by all
    order by rand_dt
    """).show()
```

```python



sql = f"""
    SELECT *
    FROM read_parquet('{f_path}')
    LIMIT 10
"""
cn.sql(sql).show()

cn.sql(f"select count(*) as row_cnt from read_parquet('{f_path}')")
```
