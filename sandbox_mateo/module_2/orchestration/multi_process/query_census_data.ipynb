```python
import duckdb

cn = duckdb.connect()

cn.execute("""
    create or replace view v_census as
    select *
    from read_json_auto('census_data/*.json')
""")

cn.sql("""
    select state_abbr, state_name, population, fips
    from v_census
    order by state_abbr
    limit 10
""").show()

cn.sql("select count(*) as total_states from v_census").show()
```
